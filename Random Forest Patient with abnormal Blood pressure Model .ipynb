```python
# importing and reading the file
abp = pd.read_csv("Patient_with_abnormal_bloodpressure.csv")
```

```python
# print shape of the data
abp.shape   # 2000 rows, 15 columns
```

```python
# check NULL values if any
abp.isnull().sum()  
```

```python
# replacing null values with binary '0' to make it absent in the respective variable

abp.Pregnancy = abp.Pregnancy.fillna(0) # 1 is for female and 0 is for male, hence feeling in all NULL's with 0
abp.Genetic_Pedigree_Coefficient = abp.Genetic_Pedigree_Coefficient.fillna(abp.Genetic_Pedigree_Coefficient.median()) # mean and median value is close so feeling the null's with median
abp.alcohol_consumption_per_day = abp.alcohol_consumption_per_day.fillna(abp.alcohol_consumption_per_day.median())  # mean and median value is close so feeling the null's with median
```

```python
# checking the records for NULLS
abp.isnull().sum()  # No NULL values present.
```

```python
# print desciption of the data set
abp = abp.iloc[: , 1:15]
abp.describe(include = 'all')  # this will print all the columns which has numeric and non numeric data
```

```python
# converting columns in category type to perform random forest
abp.Blood_Pressure_Abnormality = abp.Blood_Pressure_Abnormality.astype("category")
abp.Sex = abp.Sex.astype("category")
abp.Pregnancy = abp.Pregnancy.astype("category")
abp.Smoking = abp.Smoking.astype("category")
abp.Level_of_Stress = abp.Level_of_Stress.astype("category")
abp.Chronic_kidney_disease = abp.Chronic_kidney_disease.astype("category")
abp.Adrenal_and_thyroid_disorders = abp.Adrenal_and_thyroid_disorders.astype("category")
```

```python
# importing RNF from scikit learn
from sklearn.ensemble import RandomForestClassifier

# importing confusion matrix from scikit learn to check the data where it lies
from sklearn.metrics import confusion_matrix

# splitting data in train and test to perform the procedure

from sklearn.model_selection import train_test_split
```

```python
# splitting data into 'X' and 'Y'

abp_x = abp.iloc[: , 1:14]
abp_y = abp.iloc[: , 0]   # blood pressure abnormality is taget (Y) variable
```

```python
# splitting data into train and test 
abp_x_train, abp_x_test, abp_y_train, abp_y_test = train_test_split(abp_x, abp_y, test_size = .2, random_state = 420)
```

```python
rfc = RandomForestClassifier(n_estimators=100) # creating a RNF classifier with 100 trees 
```

```python
# fitting the model
rfc.fit(abp_x_train, abp_y_train)

```

```python
# prediction on 'x' test
pred_value = rfc.predict(abp_x_test)
pred_value
```

```python
# printing matrix on 'y' test and pred_value

mat_abp = confusion_matrix(abp_y_test, pred_value)
mat_abp
```

```python
# printing the accuracy score

from sklearn.metrics import accuracy_score
accuracy_score(abp_y_test, pred_value)
```

```python

from sklearn.metrics import recall_score   # Tp /tp +fn
recall_score(abp_y_test, pred_value)
```

```python

from sklearn.metrics import precision_score # tp / tp+fp
precision_score(abp_y_test, pred_value)
```

```python
# printing the important features score of the model variables
rfc.feature_importances_
```

```python
# converting into a list
list1 = list(rfc.feature_importances_)
list1
```

```python
# printing one to one mapping with the feature name and their Importance Score
feature_importance_rf = pd.DataFrame({"Feature_Name " : abp_x_train.columns, "Importance" : list1})
```

```python
feature_importance_rf
```

```python
# printing features in ascending order with respect to the feature's highest importance
feature_importance_rf = feature_importance_rf.sort_values("Importance", ascending = False)
feature_importance_rf
```

## Learning : 
### From the above findings we can interpret that the "Genetic_Pedigree_Coefficient" is the most significant features to trigger abnormal blood pressure in patients followed by "Level_of_Hemoglobin" and "Chronic_kidney_disease". Where "Smoking" and "Pregnancy" are the least considerable features based on   the model.

### To use the model with given data to measure individual's blood pressure with the "user define function" performing following task.

```python
from termcolor import colored  # to print in different color

def enter_input():
    Adernal_and_thyroid_disorders = eval(input("value of Adernal_and_thyroid_disorders(1: disorder, 0: Nodisorder)"))
    Age                           = eval(input("value of age::"))
    BMI                           = eval(input("value of : BMI"))
    Chronic_kidney_disease        = eval(input("value of Chronic_kidney_disease(1: for Disease, 0 : No disease)"))
    Genetic_Pedigree_Coefficient  = eval(input("value of Genetic_pedigree_Coefficient(1 : for Genetic, 0 : for not)"))
    Level_of_Hemoglobin           = eval(input("value of level_of_Hemoglobin"))
    Level_of_Stress               = eval(input("value of Level_of_Stress(1 or 2 or 3. 3 Is highest level of Stress)"))
    Physical_activity             = eval(input("value of physical_activity"))
    Pregnancy                     = eval(input("value of Pregnancy(1:if pregnant)"))
    Sex                           = eval(input("value of Sex(1 for female, 0 for male)"))
    Smoking                       = eval(input("value of Smoking(1 if smoking)"))
    alcohol_consumption_per_day   = eval(input("value of alcohol_consumption_per_day"))
    salt_content_in_the_diet      = eval(input("value of salt_content_in_the_diet"))
    df_bp                         = pd.DataFrame({"Adernal_and_thyroid_disorders" :[Adernal_and_thyroid_disorders],
                                                 "Age" : [Age],
                                                  "BMI" : [BMI],
                                                 "Chronic_kidney_disease" : [Chronic_kidney_disease],
                                                 "Genetic_Pedigree_Coefficient" : [Genetic_Pedigree_Coefficient],
                                                 "Level_of_Hemoglobin": [Level_of_Hemoglobin],
                                                 "Level_of_Stress" :[Level_of_Stress],
                                                 "Physical_activity" : [Physical_activity],
                                                 "Pregnancy" : [Pregnancy], "Sex" : [Sex],
                                                 "Smoking" : [Smoking],
                                                 "alcohol_consumption_per_day" : [alcohol_consumption_per_day],
                                                 "salt_content_in_the_diet" : [salt_content_in_the_diet],})
    pred_bp                       = rfc.predict(df_bp)
    pred_prob                     = rfc.predict_proba(df_bp)
    print(colored("-----------------------------------------------------------------", "yellow"))
    print(colored("-----------------------------------------------------------------", "blue"))
    print(colored("-----------------------------------------------------------------", "green"))
    if pred_bp ==0:
        print(colored("congrates--You do NOT have BP Abnormality--", "green"))
    else:
        print(colored("Warning --- You have Blood Pressure Abnormality --", "red"))
        print(colored("-------------------------------------------------------------", "yellow"))
        print(colored("--------------------------------------------------------------", "blue"))
        print(colored("--------------------------------------------------------------", "green"))
        print( "Your Probability of being Normal  ::: ", pred_prob[0][0] )
        print(colored("-------------------------------------------------------------", "yellow"))
        print(colored("--------------------------------------------------------------", "blue"))
        print(colored("--------------------------------------------------------------", "green"))
```
